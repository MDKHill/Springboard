```python
import pandas as pd
```

```python
df_wip = pd.read_excel('C:/Users/Hill HP 2015/Documents/Springboard/Capstone Project 1/stats.xlsx')
```

```python
# Get a feel for the data

df_wip.head()
```

```python
df_wip.info()
```

```python
# The authors of the paper didn't use missing values, so to recreate their work, I will drop rows with missing values
# However, in the interest of using all available data, I'll make a data set with the missing values intact
df_wip_with_na = df_wip
# I'll duplicate the work on this extra data set at the end of the notebook but not the head or info checks

df_wip = df_wip.dropna(axis=0, how='any')
df_wip.info()
```

```python
# Data taken from the current game and betting lines doesn't help the model, drop it

df_wip = df_wip.drop(['Opening Line', 'Closing Line', 'GF.1', 'GA.1', 'ShFor', 'ShA', 'Sh%', 'Sv%', 'Game PDO'], axis=1)
```

```python
df_wip.info()
```

```python
# Split data into home and away to rename columns for readability when concatenating back together

df_wip_home = df_wip[df_wip.location == 'Home']
df_wip_away = df_wip[df_wip.location == 'Away']
```

```python
df_wip_home.head()
```

```python
# Rename the columns to distinguish home and away data in the eventually concatenated data frame

df_wip_home.columns = ['Date','H Team','location','H Fenwick Close','H GF','H GA','H GlDiff','H PP%','H PK%','H sh%','H sv%','H PDO','H win streak','H standing','H 5-5 F/A','Final']
df_wip_away.columns = ['Date','A Team','location','A Fenwick Close','A GF','A GA','A GlDiff','A PP%','A PK%','A sh%','A sv%','A PDO','A win streak','A standing','A 5-5 F/A','Final']
```

```python
df_wip_home.head()
```

```python
# For analysis, I only need 1 column to represent home team winning, and the away data will be concatenated after
# the home data, so drop the 'Final' column from the home data, and since we know the home team data is in this
# data frame, drop the 'location' column as well

df_wip_home = df_wip_home.drop(['location','Final'], axis=1)

# The date column doesn't need to be dupliated, so drop it and the 'location' column from the away team data

df_wip_away = df_wip_away.drop(['Date','location'], axis=1)
```

```python
df_wip_home.head()
```

```python
# Make sure the home and away data have the same number of rows

df_wip_home.info()
```

```python
df_wip_away.info()
```

```python
# To make eventual index read smoothly, add 'at' before home team's name

df_wip_home['H Team'] = df_wip_home.apply(lambda row: 'at ' + row['H Team'], axis=1)
```

```python
df_wip_home.head()
```

```python
# To concatenate back together correctly, indexes of home and away data must match

df_wip_home = df_wip_home.reset_index(drop=True)
df_wip_away = df_wip_away.reset_index(drop=True)
```

```python
# Concatenate the data

df_wip_full = pd.concat([df_wip_home, df_wip_away], axis=1)
```

```python
# Create a temporary column to re-index the full data frame, set the index, drop the now unneeded columns

df_wip_full['Game'] = df_wip_full.apply(lambda row: row['Date'].strftime("%Y-%m-%d") + ' ' + row['A Team'] + ' ' + row['H Team'], axis=1)
df_wip_full = df_wip_full.set_index('Game')
df_wip_full = df_wip_full.drop(['Date', 'H Team', 'A Team'], axis=1)
```

```python
df_wip_full.head()
```

```python
df_wip_full.info()
```

```python
# Change the 'Final' column to numeric, 1 if home team won and 0 if home team lost

temp_home_win = []
for index,row in df_wip_full.iterrows():
    # Remminder, this is the away team's win or loss
    if (row['Final'] == 'Win'):
        # So if the away team won, the home team lost
        temp_home_win.append(0)
    else:
        temp_home_win.append(1)
df_wip_full['Home Win'] = temp_home_win
df_wip_full = df_wip_full.drop(['Final'], axis=1)
```

```python
df_wip_full.head()
```

```python
# Write back wrangled data set for use in capstone project

df_wip_full.to_csv('C:/Users/Hill HP 2015/Documents/Springboard/Capstone Project 1/cp1_data.csv')
```
