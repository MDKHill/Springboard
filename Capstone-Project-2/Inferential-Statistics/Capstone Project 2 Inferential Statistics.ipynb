```python
# One thing that stood out to me a little was the defensize zone starts and shot blocks for Czech players.  That makes me wonder 
# if there's a pattern to Czech player positions.  So, I might as well check (ha ha) everyone.
# But first, I need to remove duplicate player observations, so each player is only counted once.

# I'll continue using the 20 game cut-off

df_adj = df[df['GP'] > 20]

# Also, I care about the mean statistics of each individual player, but a straight mean would emphasize seasons that a
# player played more games in, so I need to "normalize" the stats that are not already percentages to per game stats 
# before going any further

stat_list = ['TOI','Goals','Total Assists','Total Points','Shots','Rush Attempts','Rebounds Created','PIM','Penalties Drawn','Giveaways','Takeaways','Hits','Hits Taken','Shots Blocked']

for stat in stat_list:
    df_adj[stat] /= df_adj['GP']

df_adj['GP'] /= df_adj['GP']
df_adj.head()
```

```python
# All right, now that all the players stats are "normalized", I'll have to write a function to create a new dataframe
# containing only mean data for all the players individually

def makeUniqueDF(inputDF):
    # Find the duplicate players
    duplicates = inputDF.duplicated(subset='Player', keep=False)
    
    # Begin the dataframe to return containing the players who already do not have multiple season data
    builtDF = inputDF[~duplicates]
    
    # Create and sort by name a dataframe of the players with multiple season data
    remainDF = inputDF[duplicates]
    remainDF = remainDF.sort_values(by='Player')
    
    # Loop until no other players remain
    while(not(remainDF.empty)):
        # Create a dataframe to store the data of the next player
        temp_player = remainDF.iloc[0]
        temp_player.iloc[7:37] = np.nan
        temp_player = pd.DataFrame(temp_player.as_matrix().reshape((1,38)))
        temp_player.columns = list(inputDF)
    
        # Get all of this player's data to work with
        this_player = remainDF.iloc[0]['Player']
        player_data = remainDF[remainDF['Player'] == this_player]
        
        # Add the temporary dataframe to the end of this player's data to put the mean data into
        player_data = pd.concat([player_data, temp_player], join='inner')
        
        # Fill the NaN's in the temporary data with the mean of the rest of the data
        player_data = player_data.fillna(player_data.mean(axis=0))
        
        # Add this player's mean data to the dataframe to return
        builtDF = pd.concat([builtDF,player_data.tail(1)], join='inner')
    
        # Remove current player from remaining players
        remainDF = remainDF[remainDF['Player'] != this_player]
    
    # Return the dataframe containing only mean player data
    return builtDF
```

```python
df_can = df_adj[df_adj['Nationality'] == 'CAN']
df_usa = df_adj[df_adj['Nationality'] == 'USA']
df_swe = df_adj[df_adj['Nationality'] == 'SWE']
df_cze = df_adj[df_adj['Nationality'] == 'CZE']
df_rus = df_adj[df_adj['Nationality'] == 'RUS']
df_fin = df_adj[df_adj['Nationality'] == 'FIN']
df_svk = df_adj[df_adj['Nationality'] == 'SVK']
df_oth = df_adj[(df_adj['Nationality'] != 'CAN') & (df_adj['Nationality'] != 'USA') & (df_adj['Nationality'] != 'SWE') & (df_adj['Nationality'] != 'CZE') & (df_adj['Nationality'] != 'RUS')& (df_adj['Nationality'] != 'FIN') & (df_adj['Nationality'] != 'SVK')]

df_can_uni = makeUniqueDF(df_can)
df_usa_uni = makeUniqueDF(df_usa)
df_swe_uni = makeUniqueDF(df_swe)
df_cze_uni = makeUniqueDF(df_cze)
df_rus_uni = makeUniqueDF(df_rus)
df_fin_uni = makeUniqueDF(df_fin)
df_svk_uni = makeUniqueDF(df_svk)
df_oth_uni = makeUniqueDF(df_oth)

# Verify the correct number of players exists is each dataframe
print('CAN {} unique players: {}'.format(len(df_can_uni),(len(df_can['Player'].unique()) == len(df_can_uni))))
print('USA {} unique players: {}'.format(len(df_usa_uni),(len(df_usa['Player'].unique()) == len(df_usa_uni))))
print('SWE {} unique players: {}'.format(len(df_swe_uni),(len(df_swe['Player'].unique()) == len(df_swe_uni))))
print('CZE {} unique players: {}'.format(len(df_cze_uni),(len(df_cze['Player'].unique()) == len(df_cze_uni))))
print('RUS {} unique players: {}'.format(len(df_rus_uni),(len(df_rus['Player'].unique()) == len(df_rus_uni))))
print('FIN {} unique players: {}'.format(len(df_fin_uni),(len(df_fin['Player'].unique()) == len(df_fin_uni))))
print('SVK {} unique players: {}'.format(len(df_svk_uni),(len(df_svk['Player'].unique()) == len(df_svk_uni))))
print('OTH {} unique players: {}'.format(len(df_oth_uni),(len(df_oth['Player'].unique()) == len(df_oth_uni))))
```

Hm, Russian players appear to have a higher mean is almost every stat, especially the production stats.  Time to run some inferences to see if that's chance or not.

```python
# First, I'll rebuild a dataframe to include all the individual non-Russian players.

df_nonrus_uni = pd.concat([df_can_uni, df_usa_uni, df_swe_uni, df_cze_uni, df_fin_uni, df_svk_uni, df_oth_uni],axis=0)

df_nonrus_uni.head()
```

```python
df_nonrus_uni.mean()
```

```python
# The primary stats players seem to be judged by are points, of course, and recently, Corsi, since puck possession can't be 
# directly measured.  Russian players also seem to have a higher mean for goals and shooting percentage, so I'll check
# those 2 as well.

import scipy
from scipy import stats

# So, null hypothesis 1: the difference between the mean points for Russian players and the mean points for other players is 0

pts_t2test_tstat, pts_t2test_pval = scipy.stats.ttest_ind(df_rus_uni['Total Points'], df_nonrus_uni['Total Points'], equal_var=False)

print('t-statistic from mean points difference test is',pts_t2test_tstat)
print('p-value from mean points difference test is',pts_t2test_pval)
```

The p-value being less than 0.05 suggests the null hypothesis can be rejected, so perhaps Russian players do score more points than other players.

```python
# So, null hypothesis 2: the difference between the mean Corsi for Russian players and the mean Corsi for other players is 0

pts_t2test_tstat, pts_t2test_pval = scipy.stats.ttest_ind(df_rus_uni['CF%'], df_nonrus_uni['CF%'], equal_var=False)

print('t-statistic from mean Corsi % difference test is',pts_t2test_tstat)
print('p-value from mean Corsi % difference test is',pts_t2test_pval)
```

This shows even stronger statistical evidence, so again, the p-value being less than 0.05 suggests the null hypothesis can be rejected, and again, perhaps Russian players are better at puck possession than other players.

```python
# So, null hypothesis 3: the difference between the mean goals for Russian players and the mean goals for other players is 0

pts_t2test_tstat, pts_t2test_pval = scipy.stats.ttest_ind(df_rus_uni['Goals'], df_nonrus_uni['Goals'], equal_var=False)

print('t-statistic from mean goals difference test is',pts_t2test_tstat)
print('p-value from mean goals difference test is',pts_t2test_pval)
```

This one's not so clean.  The p-value is less than 0.05, but not less than 0.01.  Thus, the null hypothesis can be questioned I think, but not completely rejected.  Thus, there is some statistical evidence, but not much, that perhaps Russian players score more goals than other players.

```python
# Finally, null hypothesis 4: the difference between the mean shooting percentage for Russian players and 
# the mean shooting percentage for other players is 0

pts_t2test_tstat, pts_t2test_pval = scipy.stats.ttest_ind(df_rus_uni['SH%'], df_nonrus_uni['SH%'], equal_var=False)

print('t-statistic from mean shooting percentage difference test is',pts_t2test_tstat)
print('p-value from mean shooting percentage difference test is',pts_t2test_pval)
```

Here we have a p-value greater than 0.05, so this null hypothesis cannot be rejected.
